```python
Q1='''1.The Carrizo Plains is located in which among the following countries?
a. France
b. United States
c. Italy
d. Germany'''

Q2 ='''2.Which civilization is considered as the oldest civilization of the world?
a. Indus Valley Civilization
b. Mesopotamian Civilization
c. Egyptian Civilization
d. Chinese Civilization'''

Q3='''3.Which planet has the most moons? 
a. Saturn
b. Mars
c. Jupyter
d. Uranus'''

Q4='''4.Which bird can fly straight up, down, and backward?
a. pigeon
b. Humming Bird
c. Eagle
d. Hawk'''

Q={Q1:'b' , Q2:'b' , Q3:'a', Q4:'b'}
score=0

for i in Q:
    print(i)
    
    Answer=input('\n Enter Your Answer:')
    if Answer==Q[i]:
        print(' CORRECT ANSWER!!')
        print()
        score+=1
    else:
        print('\n INCORRECT ANSWER!!')
        print()
        
print('\n Your Total Score is : ', score)
if score>4//2:
    print('\n EXCELLENT!!')
else:
    print('\n Good!!')
        


```
